By using the online simulator https://playground.tensorflow.org answer the following questions (use 20% of training data, 0 noise,  sigmoid activation function, no regularisation, and batch size equal to 10).

1. Using x1 and x2 as inputs, which of the datasets can be properly classified with one hidden layer. Why?

Three out of the 4 datasets can be successfully classified successfully:

The circular boundary can be approximated by three linear borders. A single-layer perceptron can implement one linear boundary. The use of sigmoid function helps in the approximation of curved boundaries by combining soft half-planes.



For the classification in disjoint quadrants a fourth perceptron is required



One perceptron is enough for the linearly separable problem



The spiral pattern is not successfully classified even with the maximum number of neurons (8) allowed by the simulator. It seems to provide an overtrained solution which does not seem to scale for a larger spiral
